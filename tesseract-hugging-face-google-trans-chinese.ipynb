```python
from transformers import MarianTokenizer, AutoModelForSeq2SeqLM
```

```python
text = '你跟我一起去吗？' # its will you come with me
mname = "Helsinki-NLP/opus-mt-zh-en"
tokenizer = MarianTokenizer.from_pretrained(mname)
model = AutoModelForSeq2SeqLM.from_pretrained(mname)
input_ids = tokenizer.encode(text, return_tensors="pt")
```

```python
outputs = model.generate(input_ids)
decoded = tokenizer.decode(outputs[0], skip_special_tokens=True)
print(decoded)
```

```python
def translate_chinese(text):
    mname = "Helsinki-NLP/opus-mt-zh-en"
    tokenizer = MarianTokenizer.from_pretrained(mname)
    model = AutoModelForSeq2SeqLM.from_pretrained(mname)
    input_ids = tokenizer.encode(text, return_tensors="pt")
    outputs = model.generate(input_ids)
    decoded = tokenizer.decode(outputs[0], skip_special_tokens=True)
    print(decoded)
    return decoded
```

```python

def translate_image(path):
    im = Image.open(path)
    plt.imshow(im)
    plt.show()
    custom_config = r'--oem 3 --psm 6 -l chi_sim --tessdata-dir ./'
    imgrey = im.convert('L')
    imgrey.show()
    text = pytesseract.image_to_string(imgrey, config=custom_config)
    print('extracted text: ', text)
    print('extractedtext size: ', len(text))
    
    if len(text) > 500:
        translate_chinese(text[:500])
    else: 
        translate_chinese(text)
```

```python
from googletrans import Translator
translator = Translator()

def translate_image_googletrans(path):
    im = Image.open(path)
    plt.imshow(im)
    plt.show()
    custom_config = r'--oem 3 --psm 6 -l chi_sim --tessdata-dir ./'
    imgrey = im.convert('L')
    imgrey.show()
    text = pytesseract.image_to_string(imgrey, config=custom_config)
    print('extracted text: ', text)
    print('extractedtext size: ', len(text))
    
    if len(text) > 500:
        translate_chinese(text[:500])
        print('Google translate: ')
        print(translator.translate(text[:500]))
    else: 
        translate_chinese(text)
        print('Google translate: ')
        print(translator.translate(text))
```
